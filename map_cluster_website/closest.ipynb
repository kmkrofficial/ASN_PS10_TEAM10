```python
def getDist(lat1, lon1, lat2, lon2):
    p = 0.017453292519943295
    a = 0.5 - (math.cos((lat2 - lat1) * p)/2) + (math.cos(lat1 * p) * math.cos(lat2 * p) * (1 - math.cos((lon2 - lon1) * p))/2)
    return 12742 * math.asin(math.sqrt(a))
```
